```python
import matplotlib.pyplot as plt

# Data for the pie chart
sizes = [35, 25, 25, 15]

# Create a basic pie chart
plt.pie(sizes)
plt.show()
```

```python
#Pie Chart with Labels

sizes = [35, 25, 25, 15]
labels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(sizes, labels=labels)
plt.show()
```

```python
# Data for the pie chart
sizes = [35, 25, 25, 15]
labels = ["Apples", "Bananas", "Cherries", "Dates"]
explode = [0.1, 0, 0, 0]  # Explode the first segment (Apples)

plt.pie(sizes, labels=labels, explode=explode)
plt.show()
```

```python
sizes = [35, 25, 25, 15]
labels = ["Apples", "Bananas", "Cherries", "Dates"]
colors = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode = [0.1, 0, 0, 0]  # Explode the first segment (Apples)

# Create a pie chart with custom colors, exploded segment, and a start angle
plt.pie(sizes, labels=labels, colors=colors, explode=explode, startangle=140, autopct='%1.1f%%')
plt.show()
```

```python
sizes = [35, 25, 25, 15]
labels = ["Apples", "Bananas", "Cherries", "Dates"]
explode = [0.1, 0, 0, 0]  

plt.pie(sizes, labels=labels, explode=explode, startangle=90, autopct='%1.1f%%', shadow=True) #%1.1f - atleast 1 point before decimal and .1 denotes the 
#actual number of digits after the decimal point 
plt.show()
```
